```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
bub = pickle.load(open("timebub.pkl",'rb'))
qui = pickle.load(open("timequi.pkl",'rb'))
```

```python
bubxs, bubys, quixs, quiys = list(), list(), list(), list()
for i in bub:
    bubxs.append(i[0])
    bubys.append(i[1])
for j in qui:
    quixs.append(j[0])
    quiys.append(j[1])
```

```python
plt.figure(figsize=(5,5))
example = np.linspace(100,1000)
cofactor = 1e-6*1.34
plt.xlabel("Sequence Size")
plt.ylabel('Time')
plt.title('Bubblesort')
plt.grid(True)
plt.plot(example,cofactor*example**2,linewidth=10,color='gray')
plt.plot(bubxs, bubys, color='orange')

plt.figure(figsize=(5,5))
example = np.linspace(100,1000)
cofactor = 1e-6*0.67
plt.xlabel("Sequence Size")
plt.ylabel('Time')
plt.title('Quicksort')
plt.grid(True)
plt.plot(example,cofactor*example*np.log(example),linewidth=10,color='gray')
plt.plot(quixs, quiys,color='orange')
None
```
